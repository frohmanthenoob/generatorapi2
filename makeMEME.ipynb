```python
from PIL import Image, ImageDraw, ImageFont
```

```python
def resize(image,draw,fulltext,textsize):
    width, height = image.size
    text = fulltext
    font = ImageFont.truetype('./font/msjh.ttc', textsize)
    textwidth, textheight = draw.textsize(text, font)
    margin = 5
    x = (width - textwidth - margin)/2
    y = height - textheight - margin
    return (x,y) if textwidth<=width else resize(image,draw,fulltext,textsize-1)
```

```python
def meme(filename,fulltext,textsize=64,color=(255,255,255,255)): 
    image = Image.open('./img/'+filename+'.jpg')
    width, height = image.size
    print(width,height)
    draw = ImageDraw.Draw(image)
    text = fulltext
    font = ImageFont.truetype('./font/msjh.ttc', 48)
    textwidth, textheight = draw.textsize(text, font)
    margin = 5
    x = (width - textwidth - margin)/2
    y = height - textheight - margin
    print(x,y)
    draw.text((x, y), text, font=font, fill=color)
    return image
```
